```python
import pandas as pd
import time
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import csv

# File paths
austin_path = "housing_csv_for_charts/austin.csv"
los_angeles_path = "housing_csv_for_charts/los_angeles.csv"
new_york_path = "housing_csv_for_charts/newyork.csv"

austin_data = pd.read_csv(austin_path)
los_angeles_data = pd.read_csv(los_angeles_path)
new_york_data = pd.read_csv(new_york_path)
```

```python
# Travis County housing chart
plt.rcParams["figure.figsize"] = (15,5.5)
```

```python
au_months = austin_data['Date']
au_housing = austin_data['Housing Price']

ax = plt.subplot(111)
l = ax.fill_between(au_months, au_housing)

plt.plot(au_months, au_housing)
plt.title('Travis County Housing Price',font='Times New Roman',fontsize=25,weight='bold')
plt.xlabel('Year/Month',font='Times New Roman',style='italic',fontsize=15)
plt.ylabel('Amount Per',font='Times New Roman',style='italic',fontsize=15)
plt.xticks(rotation = 90)
plt.ylim(300000,600000)

ax.spines['right'].set_color((.8,.8,.8))
ax.spines['top'].set_color((.8,.8,.8))
ax.grid('on')

plt.show()
```

```python
# Los Angeles County housing chart
la_months = los_angeles_data['Date']
la_housing = los_angeles_data['Housing Price']

ax = plt.subplot(111)
l = ax.fill_between(la_months, la_housing)

plt.plot(la_months, la_housing)
plt.title('Los Angeles County Housing Price',font='Times New Roman',fontsize=25,weight='bold')
plt.xlabel('Year/Month',font='Times New Roman',style='italic',fontsize=15)
plt.ylabel('Amount Per (Million)',font='Times New Roman',style='italic',fontsize=15)
plt.xticks(rotation = 90)
plt.ylim(800000,1050000)

ax.spines['right'].set_color((.8,.8,.8))
ax.spines['top'].set_color((.8,.8,.8))
ax.grid('on')

plt.show()
```
